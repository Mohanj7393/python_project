```python
import sqlite3
```

```python
df = pa.read_json("votes.jsonl", lines=True)
```

```python
sdb = sqlite3.connect("sql.db")
c=sdb.cursor()
print("sql db connectd")
```

```python
c.execute("CREATE TABLE Votes(Id INTEGER,PostID INTEGER,VoteTypeId INTEGER,CD DATETIME,UserId INTEGER)")
```

```python
for row in df.itertuples():
  insert_data=f"INSERT INTO Votes (Id, PostID, VoteTypeId, CD) VALUES ({row[1]}, {row[2]}, {row[3]} ,DATETIME())"
  c.execute(insert_data)
sdb.commit()
```

```python
row = c.execute("select * from votes").fetchall()
print(row)
```

```python
week1 = c.execute("""CREATE VIEW V_VOTE_WEEK1 AS
select * from votes where CD BETWEEN '1900-01-01 01:00:00' AND '2050-01-07 23:59:59'""")
print(week1)
```
